# (5.5) Determinants of the growth.
Your task is to estimate the relation $Y = f(R\&D)$ for ’your’ country (’your’ country according to exercise 5.4).

## (a)
Go to [OECD.Stat](https://stats.oecd.org/) → Science, Technology and Patents → Science and Technology Indicators → Main Science and Technology Indicators. Download the data
for GDP (current PPP\\$), population, GERD per capita population (current
PPP \\$), and BERD on R&D (current PPP \\$) for all available periods.


```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/MarkusStefan/Economics/main/OECD_Tech_Science.csv", header=0, index_col=[5])
df
```

```python
# getting the unique variables
vars = df.iloc[:, 1].unique()

# selecting only the relevant variables
vals = df.iloc[:, [1,11]]
vals
```

```python
for i, var in enumerate(vars):
  if i == 0:
    df = vals.loc[vals.iloc[:,0]==var]
    df.drop(labels="MSTI Variables", axis=1, inplace=True)
    df.columns =[var]
    continue

  var_df =  vals.loc[vals.iloc[:,0]==var]
  var_df.columns = ["",var]
  last_col = var_df.iloc[:, -1]
  df = pd.merge(df, last_col, left_index=True, right_index=True)

df
```

## (b)
Using the population data in (a) calculate GDP per capita and BERD per
capita.

```python
df['GDP per capita'] = df['Gross Domestic Product (current PPP$)'] / df['Population']
df['BERD per capita'] = df['Business Enterprise Expenditure on R&D (BERD) at current PPP $'] / df['Population']
df
```

## (c)
Estimate and discuss the relation $Y = f(R\&D)$ with GDP per capita as dependent variable and GERD per capita, BERD per capita or both as explanatory
variable(s) using OLS method. Check the consistence and significance of the
estimated parameters.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from itertools import combinations

# dependent var
Y = df['GDP per capita']

# independent vars combinations
X = [
    ['GERD per capita population (current PPP $)'],
    ['BERD per capita'],
    ['GERD per capita population (current PPP $)', 'BERD per capita']
]

for x in X:
    model = LinearRegression()
    model.fit(df[x], Y)
    beta_j = model.coef_
    beta_0 = model.intercept_
    r2 = model.score(df[x], Y)
    n = len(Y)
    p = len(x) if type(x) == list else 1
    adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

    print("Model: \t GDP per capita ~ %s" % (x[0] + ' + ' + x[1] if len(x) > 1 else x[0]))
    print("coefficients:\t %s" % ("beta1: " + str(beta_j[0]) + "; beta2: " + str(beta_j[1]) if len(beta_j) > 1 else beta_j))
    print("intercept:\t", beta_0)
    print("Adj. R2:\t %.3f\n" %adj_r2)

```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from itertools import combinations
import statsmodels.api as sm

# dependent var
Y = df['GDP per capita']

# independent vars combinations
X = [
    ['GERD per capita population (current PPP $)'],
    ['BERD per capita'],
    ['GERD per capita population (current PPP $)', 'BERD per capita']
]

for x in X:
    x_ = sm.add_constant(df[x])
    model = sm.OLS(Y, x_).fit()
    summary = model.summary()
    print(summary, "\n\n", "<>"*60, "\n", "<>"*60, "\n\n")
```

We clearly see that the CI becomes larger ones we have two iindependent variables. Also, the t-statistics of estimated parameters become smaller indicating a loss in significance. Though, all OLS models are still significant.

```python
import pandas as pd
import statsmodels.api as sm
from itertools import combinations

# dependent variable
Y = df['GDP per capita']

# independent variable combinations
X = [
    ['GERD per capita population (current PPP $)'],
    ['BERD per capita'],
    ['GERD per capita population (current PPP $)', 'BERD per capita']
]

for x in X:
    # const (beta_0)
    X_with_const = sm.add_constant(df[x])

    model = sm.OLS(Y, X_with_const)
    results = model.fit()

    t_values = results.tvalues[1:]  # discard const parameter
    p_values = results.pvalues[1:]

    # significance threshold alpha=0.05
    significant_vars = [var for var, pval in zip(x, p_values) if pval < 0.05]

    print("Model: \t GDP per capita ~ %s" % (x[0] + ' + ' + x[1] if len(x) > 1 else x[0]))
    print("Significant variables:", significant_vars)
    if len(significant_vars) < 2:
      print("t-values:\n", t_values[0])
      print("P-values:\n", p_values[0])
    else:
      print("t-values:\n", t_values[:].to_string()) # to string to get rid of dtype=float64
      print("P-values:\n", p_values[:].to_string())
    print("\n\n", "<>"*60, "\n", "<>"*60, "\n\n")
```
