**Floating (or flexible)** exchange rate regime exist where exchange rates are determined solely by market forces and often manipulated by open-market operations. Countries do have the ability to influence their floating currency from activities such as buying/selling currency reserves, changing interest rates, and through foreign trade agreements.

**Fixed (or pegged)** exchange rate regimes, exist when a country sets the value of its home currency directly proportional to the value of another currency or commodity. For years many currencies were fixed (or pegged) to gold. If the value of gold rose, the value of the currency fixed to gold would also rise. Today, many currencies are fixed (pegged) to floating currencies from major nations. Many countries fixed their currency value to the U.S. Dollar, the Euro, or the British Pound

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("https://raw.githubusercontent.com/MarkusStefan/Economics/main/Exchange_Rates_Currency_per_USD.csv", index_col=[0])
data = data.rename(columns={"Argentinia":"Argentina"})
colors = ["orange", "blue", "cyan", "red", "green", "magenta"]
data.head()


```

```python
fig, ax = plt.subplots(figsize=(15,8))

l = data.columns

for i, c in enumerate(l):
  ax.plot(data.index, data.iloc[:, i], color=colors[i])


ax.set_ylabel('exchange rate (currency / $) in %', size=14)
ax.tick_params(axis='both', labelsize=13)
ax.set_title("Exchange Rate Regimes", size=18)
ax.legend(l, prop={'size':14})
```

The Argentine peso has depreciated against the US dollar due to several reasons. One of the main reasons is the high inflation rate in Argentina. The country is battling inflation at 109%, one of the highest in the world, which a central bank poll estimates could hit nearly 130% by the end of the year
The Argentine peso has been depreciating against the US dollar for several years now. The 2018 to present Argentine monetary crisis is an ongoing severe devaluation of the Argentine peso, caused by high inflation and steep fall in the perceived value of the currency at the local level as it continually lost purchasing power, along with other domestic and international factors.


```python
fig, ax = plt.subplots(figsize=(15,8))

l = data.columns

for i, c in enumerate(l):
  if c == "Argentina":
    continue
  ax.plot(data.index, data.iloc[:, i], color=colors[i])


ax.set_ylabel('exchange rate (currency / $) in %', size=14)
ax.tick_params(axis='both', labelsize=13)
ax.set_title("Exchange Rate Regimes", size=18)
ax.legend(l[l != "Argentina"], prop={'size':14})
```

```python
fig, ax = plt.subplots(figsize=(15,8))

l = data.columns

for i, c in enumerate(l):
  if c == "Argentina" or c == "Honduras":
    continue
  ax.plot(data.index, data.iloc[:, i], color=colors[i])

mask = [True, True, False, False, True, True]
ax.set_ylabel('exchange rate (currency / $) in %', size=14)
ax.tick_params(axis='both', labelsize=13)
ax.set_title("Exchange Rate Regimes", size=18)
ax.legend(l[mask], prop={'size':14})
```

During the COVID-19 pandemic in 2020, the Brazilian currency depreciated against the US dollar. On May 13th, 2020, the Brazilian real reached a historical low against the US dollar, being negotiated at 1 USD = 5.901 BRA real. This depreciation was fueled by meager results of the economy and quick disenchantment, resulting in a lack of foreign investments.

## Discussion
While we see that fixed exchange rates remain constant over time (no surprise), we see that the exchange rates for countries following a crawling/mixed approach have tremendously depreciated in terms of USD, despite a certain level of fixation. Countries with free floating exchange rates are subject to some fluctuations, but overall, remain in a specific range.

In recent years, it was obviously beneficial to have a fixed exchange rate regime which maintained economic stability in small countries. Those countries with flexible regimes tend to be, first of all, larger and more financially stable. Countries like Argentina and Honduras with crawling regimes experienced a huge devaluation as a result of several crisis originated from not only economic but also social backgrounds.

In the period surrounding the financial crisis (Lehman Brothers) in the US in 2009, exchange rates for Brazil real and Canadian dollar decreased, meaning that the USD relatively depreciated in that period
